### Classificação Desbalanceada

- A modelagem de classificação envolve a predição de classes para uma determinada observação. 
- O problema de classificação desbalanceada é um exemplo de problema de classificação em que a distribuição de exemplos sobre as classes conhecidas não é uniforme. 
- A distribuição pode variar de um leve viés a um desbalanceamento grave, onde há um exemplo na classe minoritária por centenas, milhares, ou milhões de exemplos na maioria das classes.
- Classificações desbalanceadas representam um desafio para a modelagem preditiva, pois a maioria dos algoritmos de aprendizado usados para classificação foram projetados em torno da suposição de número de exemplos para cada classe.
- Isso resulta em modelos com baixo desempenho preditivo, especificamente para a classe minoritária. Esse é um problema porque, normalmente, a classe minoritária é mais importante e, portanto, o problema é mais sensível a erros de classificação para a classe minoritária do que para a classe majoritária. 

### Sumário
- Classificação desbalanceada como um problema de aprendizado supervisionado quando há uma distribuição desigual.
- Organização de classes no conjunto de dados de treinamento.
- A variação do desbalanceamento na distribuição de classes
    - Modelar classes disformes exige técnicas especializadas.
- Problemas de classificação no mundo real com distribuição de classe desbalanceadas:
    * Detecção de fraude;
    * Detecção de spam; 
    * Predição ou análise de churn;
    * Detecção de anomalias e eventos raros, dentre outros.

#### Exemplo 2 - código para criar dataset
- Amostra balanceada
    - 5000 exemplos por classe

```python
# criar um gráfico do dataset com classes distribuídas uniformemente
from numpy import unique
from numpy import hstack
from numpy import vstack
from numpy import where
from matplotlib import pyplot
from sklearn.datasets import make_blobs

# função para criar o dataset
def get_dataset(proportions):
	# determinar o nro de classes
	n_classes = len(proportions)
	# determinar o nro de exemplos por classe
	largest = max([v for k,v in proportions.items()])
	n_samples = largest * n_classes
	# criar dataset
	X, y = make_blobs(n_samples=n_samples, centers=n_classes, n_features=2, random_state=1, cluster_std=3)
	# criar amostras
	X_list, y_list = list(), list()
	for k,v in proportions.items():
		row_ix = where(y == k)[0]
		selected = row_ix[:v]
		X_list.append(X[selected, :])
		y_list.append(y[selected])
	return vstack(X_list), hstack(y_list)

# função que criar um scatter plot do dataset com diferentes cores para cada classe
def plot_dataset(X, y):
	n_classes = len(unique(y))
	for class_value in range(n_classes):
		# criar linhas para cada classe
		row_ix = where(y == class_value)[0]
		# plot
		pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(class_value))
	pyplot.legend()
	pyplot.show()

# definir distribuição entre as classes
proportions = {0:5000, 1:5000} # 5000 exemplos por classe
# gerar dataset
X, y = get_dataset(proportions)
plot_dataset(X, y)
print('X: %n',X) 
print('y: %n',y)
```

- O desbalanceamento de classe é geralmente medido em relação à distribuição do treinamento.
    - 10000 exemplos `0` para exemplos 1000 `1`

### Causas do Desbalanceamento entre Classes
- O desbalanceamento na distribuição de classes em uma modelagem de classificação desbalanceada pode ter muitas causas. 
- Pode-se dizer que existem dois grupos principais de causas para o desbalanceamento que se pode considerar:
    - Amostragem de dados;
    - Propriedades do domínio de um determinado problema.
- Pode ser que o desbalanceamento entre as classes de um conjunto de dados seja causado pela maneira como os exemplos foram coletados ou amostrados. Isso pode envolver vieses introduzidos durante a coleta de dados e erros cometidos durante a coleta de dados.
    - Amostragem tendenciosa;
    - Erros de mensuração.
- Erros podem ter sido cometidos ao coletar as observações. Um tipo de erro qualquer pode ter imputado classes incorretamente devido, por exemplo, a processos ou sistemas os quais as amostras coletadas foram danificadas.
- Geralmente, nos casos em que o desbalanceamento é causado por um viés de amostragem ou erro de medição, o desbalanceamento pode ser corrigido por métodos de amostragem aprimorados e/ou corrigindo o erro de mensuração.
- Muitas vezes é inviável ou intratável coletar mais amostras de um determinado domínio para melhorar a distribuição da classe. 
### - O desafio é desenvolver um modelo que aprenda as diferenças de distribuições entre as classes e mensurá-los de maneira adequada.
